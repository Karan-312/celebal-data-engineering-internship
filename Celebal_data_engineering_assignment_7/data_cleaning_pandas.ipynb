## Step 3: Handle Missing Values

### 3a. Identify missing values
We count how many missing (NaN) values exist per column.

```python
missing_counts = df.isna().sum()
missing_pct = (df.isna().mean() * 100).round(2)
missing_summary = pd.DataFrame({'missing_count': missing_counts, 'missing_pct': missing_pct})
missing_summary = missing_summary[missing_summary['missing_count'] > 0].sort_values('missing_count', ascending=False)
missing_summary
```

### 3b. Fill or drop missing values

Strategy used here (adjust to fit your own dataset/business rules):
- **Price**: numeric and important for downstream calculations → fill with the **median** price.
- **Quantity**: numeric, whole units → fill with the **median** quantity (rounded to an integer).
- **Region / Ship Mode**: categorical → fill with the **mode** (most frequent value), since dropping rows
  would lose otherwise valid data.

In general:
- Use `fillna()` when the missing data can be reasonably estimated (numeric → mean/median,
  categorical → mode) and you don't want to lose rows.
- Use `dropna()` when a critical identifier is missing (e.g., an Order ID) and the row can't be trusted.


```python
df_clean = df.copy()

# Fill numeric columns with median
for col in ['Price', 'Quantity']:
    if col in df_clean.columns:
        median_val = df_clean[col].median()
        df_clean[col] = df_clean[col].fillna(median_val)

# Round Quantity back to whole units after filling
if 'Quantity' in df_clean.columns:
    df_clean['Quantity'] = df_clean['Quantity'].round().astype(int)

# Fill categorical columns with the mode (most frequent value)
for col in ['Region', 'Ship Mode']:
    if col in df_clean.columns:
        mode_val = df_clean[col].mode(dropna=True)[0]
        df_clean[col] = df_clean[col].fillna(mode_val)

# Example of dropna(): drop any row still missing a critical identifier (Order ID)
df_clean = df_clean.dropna(subset=['Order ID'])

print("Remaining missing values after cleaning:")
df_clean.isna().sum().sum()
```

## Step 4: Basic Operations — Filter Rows & Select Columns

```python
# Select specific columns
subset_cols = df_clean[['Order ID', 'Category', 'Price', 'Quantity']]
subset_cols.head()
```

```python
# Filter rows: e.g. orders in the 'Technology' category with Price > 100
filtered = df_clean[(df_clean['Category'] == 'Technology') & (df_clean['Price'] > 100)]
print(f"Rows matching filter: {len(filtered)}")
filtered.head()
```

```python
# Filter rows: orders with quantity of 5 or more
bulk_orders = df_clean[df_clean['Quantity'] >= 5]
print(f"Bulk orders (Quantity >= 5): {len(bulk_orders)}")
bulk_orders.head()
```

## Step 5: Remove Duplicates

```python
num_duplicates = df_clean.duplicated().sum()
print(f"Number of fully duplicated rows: {num_duplicates}")

df_clean = df_clean.drop_duplicates().reset_index(drop=True)
print(f"Shape after removing duplicates: {df_clean.shape}")
```

## Step 6: Create a Derived Column

`total_amount = price * quantity`

```python
df_clean['total_amount'] = df_clean['Price'] * df_clean['Quantity']
df_clean[['Order ID', 'Price', 'Quantity', 'total_amount']].head()
```

```python
print("Total sales across all cleaned orders: ${:,.2f}".format(df_clean['total_amount'].sum()))
print("Average order amount: ${:,.2f}".format(df_clean['total_amount'].mean()))
```

## Step 7: Save the Cleaned Dataset

```python
output_path = "superstore_cleaned.csv"
df_clean.to_csv(output_path, index=False)
print(f"Cleaned dataset saved to: {output_path}")
print(f"Final shape: {df_clean.shape}")
```

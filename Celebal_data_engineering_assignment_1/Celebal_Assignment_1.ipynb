## Handling Missing Values

```python

df.isnull().sum()
```

```python
df['discount'] = df['discount'].fillna(0)
```

```python
text_fill_cols = [
    'what_customers_said',
    'seller_name',
    'videos',
    'variations',
    'seller_information'
]

for col in text_fill_cols:
    df[col] = df[col].fillna("Not Available")
```

## Filtering and Selecting Data

```python
premium_items = df[df['initial_price'] > 5000]

premium_items.head()
```

```python
subset_df = df[['product_id', 'title', 'category', 'initial_price', 'rating']]

subset_df.head()
```

## Remove Duplicates

```python
df.duplicated(subset=['product_id']).sum()
```

```python
df = df.drop_duplicates(subset=['product_id'])

df.shape
```

 ## Creating New Columns

```python
df['price'] = df['initial_price'].astype(float)
```

```python
np.random.seed(42)

df['quantity'] = np.random.randint(1, 5, size=len(df))
```

```python
df['total_amount'] = df['price'] * df['quantity']
```
